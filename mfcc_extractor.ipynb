```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
import os
import tensorflow as tf
warnings.filterwarnings('ignore')
DATA_PATH = '../birdsong-recognition/train_audio/'
```

```python
#Reading train.csv
df = pd.read_csv(r"D:\\Project\\Main\\birdsong-recognition\\train.csv")
df.head() 
```

```python
def extract_mfcc(filename,bird_name):
    path = os.path.join(DATA_PATH,bird_name,filename)
    song,sr = librosa.load(path)
    mfcc = librosa.feature.mfcc(song,sr=sr,n_mfcc=40).T
    
    return mfcc
```

```python
df = df.query("rating>=5")
df.head()
```

```python
df.shape[0]
```

```python
data = {
    "unique labels":[],
    "labels":[],
    "mfccs":[]
}
```

```python
unique_labels = {}
count = 0
for index, row in df.iterrows():
    filename = row["filename"]
    bird_name = row["ebird_code"]
    mfcc = extract_mfcc(filename,bird_name)
    index = unique_labels.get(bird_name)
    if index == None:
        data["unique labels"].append(bird_name)
        index = len(data["unique labels"])-1
        unique_labels[bird_name] = index
    data["labels"].append(index)
    data["mfccs"].append(mfcc.tolist())
    count+=1
    if count == 4500:
        break
    if count%100 == 0:
        print(count)
print(count)
```

```python
print("Number of mfccs saved: {}".format(len(data["mfccs"])))
print("Number of labels saved: {}".format(len(data["labels"])))
```

```python
import json
json_path = "data.json"
with open(json_path, "w") as fj:
    json.dump(data, fj, indent=4)
```
